```python
cr1 = ClassicalRegister(3, "bytes")
qr1 = QuantumRegister(3, "qubytes")
extrar = QuantumRegister(1, "extraq")
msc = QuantumCircuit(qr1, extrar, cr1)
msc.h(qr1[2])
msc.h(qr1[0])
msc.cx(extrar, qr1[0])
msc.cx(qr1[1], extrar)
msc.measure(qr1, cr1)
msc.draw()
```

```python
backend = BasicAer.get_backend('qasm_simulator')
job = execute(msc, backend)
plot_histogram(job.result().get_counts(), color='midnightblue', title="New Histogram")
```

```python
qubitreg = QuantumRegister(3, 'qubytes')
newtreg = ClassicalRegister(3, 'regbits')
extranewts = ClassicalRegister(1, 'extraqubit')
nextqc = QuantumCircuit(qubitreg, newtreg, extrar, extranewts)
nextqc.rx(np.pi/2, extrar)
nextqc.sx(extrar)
nextqc.rz(np.pi/2, extrar)
nextqc.cx(extrar, qubitreg[2])
nextqc.rz(np.pi/4, extrar)
nextqc.cx(extrar, qubitreg[1])
nextqc.rz(-np.pi/4, extrar)
nextqc.cx(extrar, qubitreg[0])
nextqc.rz(np.pi/4, extrar)
nextqc.cx(extrar, qubitreg[1])
nextqc.rz(-np.pi/4, extrar)
nextqc.cx(extrar, qubitreg[2])
nextqc.rz(np.pi/4, extrar)
nextqc.cx(extrar, qubitreg[1])
nextqc.rz(-np.pi/4, extrar)
nextqc.cx(extrar, qubitreg[0])
nextqc.rz(np.pi/4, extrar)
nextqc.rz(np.pi/4, qubitreg[2])
nextqc.rz(np.pi/4, qubitreg[2])
nextqc.rz(np.pi/4, qubitreg[1])
nextqc.rz(np.pi/2, extrar)
nextqc.sx(extrar)
nextqc.rz(np.pi/2, extrar)
nextqc.cx(qubitreg[2],qubitreg[0])
nextqc.cx(qubitreg[1], qubitreg[0])
nextqc.rz(np.pi/4, qubitreg[0])
nextqc.rz(-np.pi/4, qubitreg[1])
nextqc.rz(-np.pi/4, qubitreg[2])
nextqc.cx(qubitreg[2],qubitreg[0])
nextqc.cx(qubitreg[1], qubitreg[0])
nextqc.measure(extrar, extranewts)
nextqc.measure(qubitreg, newtreg)
nextqc.draw(output="latex")
```

```python
thetoffjob = execute(nextqc, backend)
```

```python
plot_histogram(thetoffjob.result().get_counts(), color="midnightblue")
```
