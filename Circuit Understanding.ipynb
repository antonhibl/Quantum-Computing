```python
import numpy as np
import pandas as pd
import matplotlib as mpl
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, execute, transpile, Aer, BasicAer, IBMQ, QuantumRegister, ClassicalRegister
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
# Creating a circuit and it's registers
qubreg = QuantumRegister(2, 'qubit')   # Defines the quantum register
bitreg = ClassicalRegister(2, 'bit')   # Defines the classical register
epicircius = QuantumCircuit(qubreg, bitreg)   # Defines a circuit called epicircius

# Instructions for the circuit below
epicircius.h(qubreg[0])
epicircius.cx(qubreg[0], qubreg[1])
epicircius.measure(qubreg, bitreg)
epicircius.draw(output='latex')
```

```python
# Defining a backend and respective job using the qasm simulator from qiskit
backend = BasicAer.get_backend('qasm_simulator')   # Using basic aer to deifne the backend
epicjob = execute(epicircius, backend)   # defines the job itself, using the circuit and backend
```

```python
plot_histogram(epicjob.result().get_counts(), color='midnightblue', title='epic plot')
```
